```python
from matplotlib import pyplot as plt
x=[12,33,45]
y=[7,9,18]

x2=[6,9,11]
y2=[6,15,7]
plt.plot(x,y,'g',label='line one',linewidth=5)
plt.plot(x2,y2,'c',label='line two',linewidth=5)
plt.title('info')
plt.ylabel('y axis')
plt.xlabel('x axis')
plt.legend()
plt.show()
```
